# Support Vector Machine
SVM classifier as fitting the widest possible street between the classes(large margin classification)
- SVMs are sensitive to the feature scales, so you should use **Standarization** (StandardScale)
    - because LinearSVC regularize the bias term(so data shoud be substracted by the mean)
- For Linear & Non linear data
- SVM classifiers do not out‚Äêput probabilities for each class.

### Tuning
- C is a hyperparameter 
    - low C = wider street & more violations(let some wrong instances not affect the decision boundary)
    - high C = narrow street & low violations(strict to fit the instances)
   
- If your SVM model is overfitting, you can try regularizing it by
reducing C.


```python
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
import numpy as np

def plot_all(clf, X, y, axes):
    plot_svm_boundary(clf, axes)
    plot_dataset_samples(X,y,axes)

def plot_svm_boundary(clf, axes):
    x0s = np.linspace(axes[0], axes[1], 100)
    x1s = np.linspace(axes[2], axes[3], 100)
    x0, x1 = np.meshgrid(x0s, x1s)
    X = np.c_[x0.ravel(), x1.ravel()]
    y_pred = clf.predict(X).reshape(x0.shape)
    y_decision = clf.decision_function(X).reshape(x0.shape)
    plt.contourf(x0, x1, y_pred, cmap=plt.cm.brg, alpha=0.2)
    plt.contourf(x0, x1, y_decision, cmap=plt.cm.brg, alpha=0.1)
    

def plot_dataset_samples(X, y, axes):
    plt.plot(X[:, 0][y==0], X[:, 1][y==0], "bs")
    plt.plot(X[:, 0][y==1], X[:, 1][y==1], "g^")
    plt.axis(axes)
    plt.grid(True, which='both')
    plt.xlabel(r"$x_1$", fontsize=20)
    plt.ylabel(r"$x_2$", fontsize=20, rotation=0)
```

```python
from sklearn.datasets import make_moons
X, y = make_moons(n_samples=100, noise=0.15, random_state=42)
```

```python
X.shape, y.shape, X[0], y[0]
```

# Non Linear Data (Polynomial data)

```python
axes = [-1.5, 2.5, -1, 1.5]
plot_dataset_samples(X,y,axes)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.svm import LinearSVC
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import PolynomialFeatures
```

## Adding Polynomial Features with Linear SVM

```python
svm_poly = Pipeline([
    ("poly", PolynomialFeatures(degree=3)),
    ("std", StandardScaler()),
    ("svm", LinearSVC(C=10, loss="hinge", max_iter=10000))
])
```

```python
svm_poly.fit(X, y)
```

```python
plot_all(svm_poly,X,y,axes)
```

## SVC Poly Kernel

```python
from sklearn.svm import SVC
```

```python
# specify the degree of polynomial featres
# bigger degree = more overfitting
# bigger C = more overfitting
# coef1 = r = controls how much the model is influenced by highdegree polynomials versus low-degree polynomials
svm_kernel_poly = Pipeline([
    ("std", StandardScaler()),
    ("svm", SVC(kernel="poly", degree=3, C=10, coef0=1))
])
```

```python
svm_kernel_poly.fit(X,y)
```

```python
plot_all(svm_kernel_poly,X,y,axes)
```

## SVC Gaussian RBF Kernel
Increasing
gamma makes the bell-shape curve narrower, and as a
result each instance‚Äôs range of influence is smaller: the decision boundary ends up
being more irregular, wiggling around individual instances. Conversely, a small gamma
value makes the bell-shaped curve wider, so instances have a larger range of influ‚Äê
ence, and the decision boundary ends up smoother. So Œ≥ acts like a regularization
hyperparameter: if your model is overfitting, you should reduce it, and if it is under‚Äê
fitting, you should increase it (similar to the C hyperparameter)

- Suitable for small/medium datasets since it is very complex

```python
svm_rbf = Pipeline([
    ("std", StandardScaler()),
    ("svm", SVC(kernel="rbf", C=1, gamma=0.5))
])
```

```python
svm_rbf.fit(X,y)
```

```python
# you will see more violations as C (the weight of the violations in the loss function is low)
plot_all(svm_rbf,X,y,axes)
```

```python
# increasing C (the weight of the violations in the loss function) will make the model has not violations (overfit)
svm_rbf = Pipeline([
    ("std", StandardScaler()),
    ("svm", SVC(kernel="rbf", C=100, gamma=0.5))
])
svm_rbf.fit(X,y)
plot_all(svm_rbf,X,y,axes)
```

```python
# increasing gamma will make the model 
svm_rbf = Pipeline([
    ("std", StandardScaler()),
    ("svm", SVC(kernel="rbf", C=100, gamma=5))
])
svm_rbf.fit(X,y)
plot_all(svm_rbf,X,y,axes)
```
